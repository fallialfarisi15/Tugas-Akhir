```python
def remove_pattern(text, pattern_regex):
    r = re.findall(pattern_regex, text)
    for i in r:
        text = re.sub(i, '', text)
    return text
```

```python
tweet_df['clean_text']= np.vectorize(remove_pattern)(tweet_df['text'],"https?://\S+|www\.\S+ | *@\w+")
tweet_df.head(1050)
```
